# Analysis Notebook Generator

Tool to initialize analysis notebooks for quick processing raw calibration data. 

To generate some analysis notebook simply run the command ```pm.tools.cal.db_analysis_generator()``` in the cell below. 
- The resulting analysis notebooks will be datestamped and generated in the current folder.
- The resulting notebooks can be accessed and executed from the jupyter notebook navagator used to open this manager. 

You can specify the relative location you want the notebook generated by calling:   
```pm.tools.cal.db_analysis_generator(loc = './YOURNOTEBOOKPLACE/')```

**Always remember to log the user when runnning the generated notebooks, and print the notebooks to .html when complete.**
